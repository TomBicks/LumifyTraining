## Introducing Lists

## Performing Useful List Operations

```python
# When adding lists together, just like with numbers we can use e.g. += to do addition and equals in one line
my_list1 = [12, 24]
my_list2 = [36, 48]
```

```python
# We can reverse the current order of a list by using .reverse
my_list1.reverse()
my_list1
```

## Perform Slicing Operations on Lists

```python
# List slicing operations allow us to access a range of elements, rather than just a single one
my_car_list = ['Honda', 'Kia', 'Nissan', 'Toyota']
```

```python
# By specifying a range of indexes, using [x:y], we use our first slicing operations
my_car_list[0:3]
# Notice despite going up to index 3, 'Toyota' wasn't included? This is because the start index is inclusive, but the end index is exclusive.
```

```python
my_car_list[1:2]
```

```python
# If we don't specify the start index, it will default to 0
my_car_list[:3]
```

```python
# If we don't specify the end index, it will default to the last index of the list
my_car_list[2:]
```

```python
# If we store the results of a slice operations, we've actually created a deep copy, and changes to this new list will not affect the original
sliced_car_list = my_car_list[0:2]
sliced_car_list
```

```python
sliced_car_list.remove('Kia')
print('sliced_car_list is', sliced_car_list)
print('car_list is', my_car_list)
```

```python
# Negative indexes also work (remember, the end index is still exclusive, but also with negatives doesn't start from 0)
# my_car_list = [0->'Honda', 1->'Kia', 2->'Nissan', 3->'Toyota']
# my_car_list = [-4->'Honda', -3->'Kia', -2->'Nissan', -1->'Toyota']
# Here, we're starting at index 0, then going to -2, which as we know will, counting backwards, exclude the 2nd to last element, 'Nissan', onwards
my_car_list[0:-2]
```

```python
# All negative indexes work too; if we want to go from the 3rd to last element, 'Kia', up to the 2nd element, 'Nissan', we can use [-3:-1]
# NOTE!! Remember! Again, the last index is *not* inclusive! This may refer to the last element, but it does not *include* it
my_car_list[-3:-1]
```

```python
# If we use a range that makes no sense, such as the start index after the last index, python returns an empty list
my_car_list[3:2]
```

## Using Step Size in Slicing Operations

```python
# The step size tells python how we want to access a range of elements between the start and end indexes
# By default, python assumes a step size of 1, meaning each element will be access one after the other
step_car_list = ['Honda', 'Kia', 'Nissan', 'Toyota', 'Ferarri']
```

```python
# To specify the step size, we state it after the range, where x is the start index, y is the end index, the step size is z [x:y:z]
step_car_list[1:4:1]
```

```python
# However, just like how we can omit the start index and the end index, we can omit both, assuming start and end, to just specify step size
step_car_list[::1]
```

```python
# If we specify a step size of 2, we'll skip every 2nd element, incrementing index two at a time
step_car_list[1:4:2]
# Thus we skip 'Nissan'
```

```python
# Using a negative step size, we can step through the list in a reverse order
step_car_list[::-1]
```

```python
step_car_list[::-2]
```

```python
# Last topic, we said a start index higher than an end index returned an empty array, meaning python could make no sense of it
# If the step size is negative however, it CAN; here we start at index 3, then work backwards to index 0 (non-inclusive)
step_car_list[3:0:-1]
```

```python
# This strange assignment allows us to assign new elements to a list, though is unorthodox
step_car_list[5:-1] = ['Lamboughini', 'Tesla']
step_car_list
```

```python
# As we know, the pop function will pop an element at the specified index; these two work with negatives
step_car_list.pop(-3)
# 'Ferarri' is removed; the 3rd last element
step_car_list
```

## Introducing Tuples

```python
# A tuple is a complex data type commonly used in Python; similar in many ways to lists, but also different
# Similarly to lists, tuples are an ordered collection of elements; the order of elements matters
```

```python
# We can initialise a tuple using rounded brackets
my_tuple = ()
my_tuple
```

```python
type(my_tuple)
```

```python
# Thus, the only difference in initialisation so far between a list and a tuple is the brackets use
int_tuple = (1, 2, 3)
int_tuple
```

```python
string_tuple = ("Hello", "Python")
string_tuple
```

```python
# They can also contain multiple different data types
combined_tuple = int_tuple + string_tuple
combined_tuple
```

```python
# They support multiplcation too, like lists
combined_tuple * 2
```

```python
# Any time you assign a value to a variable, where the values are seperated by commas, Python assumes it's a tuple
my_tuple1 = (1, "Hello", 3.4)
my_tuple2 = 1, "Hello", 3.4
print(my_tuple1)
print(my_tuple2)
```

```python
# They can also be destructured the same as lists (we refer to elements within a tuple as fields)
a, b, c = my_tuple1
print(a)
print(b)
print(c)
```

```python
# Tuples can also be nested
nested_tuple = (1, 2, 3, (4, 5, 6))
nested_tuple
```

```python
# And nested tuples can be very mixed
mixed_nested_tuple = ('mouse', (8, 4, 6), (1, 2, 3))
mixed_nested_tuple
```

```python
a, b, c = mixed_nested_tuple
print(a)
print(b)
print(c)
```

```python
# We access tuple's fields the same we would with lists; using '[]'
mixed_nested_tuple[1]
```

```python
# To access the field in a nested tuple, we use multiple sets of '[]'
mixed_nested_tuple[1][1]
```

```python
# They also support slicing operations
nested_tuple[1:-1]
```

```python
# Python also supports the 'in' keyword for both tuples and lists, to return whether a value is present within a tuple or list
2 in nested_tuple
```

```python
# nested_tuple = (1, 2, 3, (4, 5, 6))
# This won't get into the nested fields however, because those are of data type tuple
5 in nested_tuple
```

## Understanding Tuple Immutability

```python
# We've gone on about what makes lists and tuples similar; what makes them different?
# The primary difference is that tuples, once created, *cannot* be updated (thus being immutable)
my_tuple = (3, 6, 9)
```

```python
my_tuple[1] = 3
```

```python
del my_tuple[1]
```

```python
# Whilst we can't delete a field (again, immutable), we can still delete the tuple as a whole
del my_tuple
my_tuple
```

```python
# However, whilsts tuples are immutable, nested mutable fields such as lists nested within tuples are now
my_tuple2 = (6, 9, 55, [22, 50])
my_tuple2
```

```python
my_tuple2[3]
```

```python
my_tuple2[3][1] = 17
my_tuple2
```

```python
# We can perform interesting functions on tuples (and also at least lists), such as the .zip function
tuple_a = (1, 2, 3, 4, 5)
tuple_b = ('a', 'b', 'c', 'd', 'e')
```

```python
# The zip function essentially makes pairs of every element in tuple a with every element in tuple b, with the result being an object
zipped = zip(tuple_a, tuple_b)
zipped
# In its simplest form, the zip data structure can be thought of as a collection of tuples, each having one field from tuple+a and one from tuple_b
```

```python
# What exactly the collection is depends on how you use the zipped result
# The zipped result can be converted to any complex data type that python supports
# If we pass the zipped result to a tuple function, we'll end up with a tuple of tuples
result = tuple(zipped)
print(result)
# As we can see, each field from each tuple we combined into tuples
```

```python
# As mentioned briefly before, we can use .zip on lists as well
list_a = [6, 7, 8, 9]
list_b = ['a', 'b', 'c', 'Hi']
```

```python
zipped_list = zip(list_a, list_b)
zipped_list
```

```python
# Notice that despite this working with lists, the elements inside are all tuples; this is how the .zip function works
result = list(zipped_list)
print(result)
```

```python
# Here is an example where we convert the results into a dictionary
a = ("John", "Charles", "Mike")
b = ("Manager", "Supervisor", "Engineer")
```

```python
x = zip(a, b)
x
```

```python
print(dict(x))
```
