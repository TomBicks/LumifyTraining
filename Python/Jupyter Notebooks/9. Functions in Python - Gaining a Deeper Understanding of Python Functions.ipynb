## Global and Local Variables

```python
# These are global variables, defined outside any functions
country = "AUS"
city = "Sydney"
```

```python
# This function directly references the global scope variables
def some_fn():
    print("Country", country)
    print("City", city)
```

```python
some_fn()
```

```python
# Local variables are those defined within a function; they cannot be accessed from outside the function
# They also take precedence over global variables of the same name within the scope; take this local variable also named country
def some_fn():
    country = "India"
    
    print("Country", country)
    print("City", city)
```

```python
some_fn()
```

```python
# The global variables remain untouched, despite the local variable being changed
country, city
```

```python
# This means we can't assign a new value to the global country, as "country = value" just creates a local variable, as shown here
def some_fn():
    # Here, we've actually referred to a local variable, but before we assign a value to it, causing an error
    country = country + "TRALIA"
    
    print("Country", country)
    print("City", city)
```

```python
some_fn()
```

```python
# However, if we assign a value to this local variable first, then it works, but it still won't reference the global variable, or change it
def some_fn():
    country = "OZ-"
    country = country + "TRALIA"
    
    print("Country", country)
    print("City", city)
```

```python
some_fn()
```

```python
city, country
```

```python
# So what do we do if want want to refer to the global variable specifically? We use the global keyword
# TODO!!
def some_fn():
    global city, country # We declate the variable as global, which means any reference from here on to city or country will be to the global variable
    print(city, country)

    city += "-SYD"
    country += "TRALIA"
    
    print("Country", country)
    print("City", city)
```

```python
some_fn()
```

```python
city, country
# This time, our changing of the variables *did* affect our global variables, because we specifically referenced them in the function
```

## Arguments

### Passing by Value

```python
# We can define a function two different ways to reflect how one can refer to global or local variables
```

```python
salary = 1000
expense = 200

name = "James"
interest_rate = 0.4
```

```python
# Here we've defined a function with no local variables, which means all these variable names will look to global variables for values
def print_savings_details():
    print("Name on account:", name)

    savings = salary - expense
    print("Savings:", savings)

    print("Interest rate:", interest_rate)
```

```python
print_savings_details()
```

```python
# Here, instead we've given arguments to the function, which means these variable names now refer to the local variables (our arguments)
def print_savings_details(name, salary, expense, interest_rate):
    print("Name on account:", name)

    savings = salary - expense
    print("Savings:", savings)

    print("Interest rate:", interest_rate)
```

```python
print_savings_details()
```

```python
print_savings_details("James", 1000, 200, 0.4)
```

```python
print_savings_details("Julie", 2000, 1600, 0.6)
```

```python
# Here, using the same function as before (with arguments) we pass in the value of the global variables for the function's local variables
print_savings_details(name, salary, expense, interest_rate)
```

```python
# But what if we also redefine the function to increase expenses? Will it affect the global variable too?
def print_savings_details(name, salary, expense, interest_rate):
    print("Name on account:", name)

    expense += 100
    savings = salary - expense
    print("Savings:", savings)

    print("Interest rate:", interest_rate)
```

```python
print_savings_details(name, salary, expense, interest_rate)
# We see expenses has increased, but has the global variable changed?
```

```python
name, salary, expense, interest_rate
# The answer is no; whilst we passed in the value of the global variable, it was not a direct reference to it; changes were not reflected
# In other words, this is what is meant when we "Pass by Value"
```

```python
# The same is true of complex data; if we try to change the list we pass in by value, it won't update the global variable
fruits_list = ["Apple", "Grapes", "Mango", "Bananas" ]

def change_list(fruits_list):
    fruits_list = ["Kiwi", "Strawberry"]
    print("Inside the function: ", fruits_list)
```

```python
change_list(fruits_list)
print("Outside the function: ", fruits_list)
# As we can see, no change to the global variable; we had no real reference to it, only its value
```

### Passing by Reference

```python
# Whilst changing the value of variables passed into a function as an argument was shown to not change the original, what about complex data types?
fruits_list = ["Apple", "Grapes", "Mango", "Bananas"]
```

```python
def change_list(fruits_list):
    # Will this change the value in the original list?
    fruits_list[0] = "Kiwi"
    print("Inside the function:", fruits_list)
change_list(fruits_list)
print("Outside the function:", fruits_list)
# We see that Apple has been replaced both within the function and without; This means the fruits-list variable we changed was the original, not a copy
```

```python
fruits_list = ["Apple", "Grapes"]
change_list(fruits_list)
print("Outside the function:", fruits_list)
```

**A good explanation of why this list changes but the strings and integers we've passed before didn't change can be found on [Stack Overflow](https://stackoverflow.com/a/17686659).
The basic gist however is that with the list, we are mutating the contents of the list directly, rather than assigning it to a different varaible, whilst in, say a function to subtract a given value by 2, we're assign our value to a new, local variable, then returning the result.**

```python
fruits_list = ["Apple", "Grapes", "Mango", "Bananas"]

# Despite the above however, even if we assign the list to another variable, then alter the list from said variable, it still changes the original list
def change_list(fruits_list):
    # Will this change the value in the original list?
    inside_list = fruits_list
    inside_list[0] = "Kiwi"
    print("Inside the function:", inside_list)
change_list(fruits_list)
print("Outside the function:", fruits_list)
```

**Referring to our earlier paragraph, the reason one data type ends up only passed by value to a function, whilst another ends up passed by reference, is that not all data types behave the same when passed in as arguments.
The reason for this is mutability. 
If we pass in mutable data types; those that can be changed after being created, like lists and dictionaries, they pass by reference and any changes made to them in the function will be reflected outside the function, because they refer to the same object.
If we pass in immutable data types; those that cannot be changed after being created, like tuples, strings, ints, floats, etc., they pass by value and, whilst their value; the object, can be assigned to a variable, it never refers back to the original, and changes made will *not* affect the original object.
More reading on Immutable vs Mutable Data Types can be found [here](https://www.geeksforgeeks.org/python/mutable-vs-immutable-objects-in-python/).**

```python
# As explained, not all complex data types are mutable; tuples are not, for instance. But dictionaries are, as well as lists
car_speeds_dictionary = {
    "Camry": 120,
    "Accord": 130,
    "Cayenne": 250,
    "Mini Cooper": 100
}
```

```python
def change_dictionary(car_speeds_dictionary):
    car_speeds_dictionary["Mustang"] = 200
    print("Inside the function:", car_speeds_dictionary)
change_dictionary(car_speeds_dictionary)
print("Outside the function:", car_speeds_dictionary)
```

```python
# We see the changes we made to the dictionary inside the function have taken affect outside the function; because it was passed in by reference
car_speeds_dictionary
```

```python
# Even if we refer to it by another name within the function, it's still a reference to the original dictionary
def change_dictionary(some_dictionary):
    some_dictionary["Camry"] = 140
    print("Inside the function:", some_dictionary)
change_dictionary(car_speeds_dictionary)
print("Outside the function:", car_speeds_dictionary)
```

```python
# And if we delete an entry from the dictionary in function, it will be reflected on the original too
def change_dictionary(car_speeds_dictionary):
    del car_speeds_dictionary["Camry"]
    print("Inside the function:", car_speeds_dictionary)
change_dictionary(car_speeds_dictionary)
print("Outside the function:", car_speeds_dictionary)
```

```python
# If we define a function without a name, we've created an anonymous function; a lambda
# Focussing on the right hand side of this lambda, which cubes a value, we have; 
# 1. lambda, indicating it's an anonymous function
# 2. x, the input argument (in this example, the only input argument; we can define more)
# 3. lastly we have the expression (which is constrained and can't have complex statements; for example, there can't be if else statements)
cube_of = lambda x: x * x * x
# We've then assigned this lambda function to a variable, cube_of
```

```python
# Here we have a lambda with two input arguments; lets recreate the calc functions
add = lambda x, y : x + y

sub = lambda x, y : x - y

mul = lambda x, y : x * y

div = lambda x, y : x / y
```

### Filter Function

```python
# We've seen how lambdas can get very complicated and hard to read; the question then is; why would we use them over regular functions?
```

```python
# Lambdas are often used alongside the build-in filter function in Python
```

```python
num_list = [1, 5, 6, 7, 11, 78, 99, 34, 105, 214]
```

```python
# The filter function takes two arguments; the function object, in this case the lambda, and the list we want to filter
filter_object = filter(lambda x : x > 10, num_list)
filter_object
# Note it returns a filter object; we can't make sense of it right now
```

```python
# If we convert it to a list, we see what it has returned
list(filter_object)
```

```python
# The logic behind filter() is that the lambda is applied to every element in the list, returning true or false for each based on the results
greater_than_10_list = list(filter(lambda x : x > 10, num_list))
greater_than_10_list
# We see that the filter function has returned only the elements of num_list which matched the condition, passing them to the new list
```

```python
even_list = list(filter(lambda x : x % 2 == 0, num_list))
even_list
```

```python
greater_than_10_less_than_100_list = list(filter(lambda x : x > 10 and x < 100, num_list))
greater_than_10_less_than_100_list
```
