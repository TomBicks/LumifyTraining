## If-else Elif Statements

```python
if 10 > 20:
    print("10 is greater than 20!")
    print("If block activated")
else:
    print("10 is less than 20!")
    print("Else block activated")
```

```python
num = 50
```

```python
if num < 100:
    num += 50
else:
    num -=50
```

```python
num
```

```python
# We can use the elif (else if) statement to make multiple if checks in a single if statement
if 15 > 20:
    print("15 is greater than 20!")
    print("If block activated")
elif 15 < 20:
    print("15 is less than 20!")
    print("Elif block activated")
else:
    print("15 is equal than 20!")
    print("Else block activated")
```

## Nested if-else Statements

```python
# We can have as many elif statements as we like
bike_price = 100
if bike_price <= 10:
    print("Bike is cheap")
    print("If block activated")
elif bike_price <= 50:
    print("Bike is moderate")
    print("1st Elif block activated")
elif bike_price <= 100:
    print("Bike is pricey")
    print("2nd Elif block activated")
elif bike_price <= 150:
    print("Bike is *very* pricey")
    print("3rd Elif block activated")
else:
    print("Bike is *wildly* expensive!")
    print("Else block activated")
```

```python
# We can also nest if statements within if statements
# If else statements are paired using additional indentation from the left relative to lines just before and after the block
age = int(input("How old are you?")) # Don't forget to convert the age to an int
if age > 12:
    print("person is older than a child")
    if age > 18:
        print("person is an adult")
    else:
        print("Person is a teenager")
else:
    print("person is a child")
```

## If-else Statements with Complex Datatypes

### Lists

```python
fruit_list = ['Grape', 'Apple', 'Banana', 'Orange']
if 'Grape' in fruit_list:
    print("Found a Grape!")
else:
    print("No Grapes...")
```

```python
if fruit_list[1] == 'Apple':
    print("Found an Apple at index 1!")
```

```python
if fruit_list[3] != 'Blackberry':
    print(fruit_list)
    fruit_list[3] = 'Blackberry'
    print(fruit_list)
```

### Tuples

```python
car_tuple = ('Toyota', 'Kia', 'Honda')
if 'Ferarri' in car_tuple:
    print("My Ferarri!")
else:
    print("No Ferarri...")
```

### Dictionaries

```python
salary_details = {"Lisa":25000,
                  "Jason":45000,
                  "Cooper":35000,
                  "Elias":23000,
                  "Jordan":77000}
```

```python
if 'Cooper' in salary_details:
    print("Found Cooper's salary details")
```

```python
# We can also check for values by specifying for values
if 77000 in salary_details.values():
    print("Someone is being paid 77,000")
else:
    print("No one is being paid 77,000")
```
