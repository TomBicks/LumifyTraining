## TODO!! Return Values

### Functions

```python
# Here, we expect the substraction to be returned:
def substract(num_1, num_2):
    result = num_1 - num_2
```

```python
value = substract(10, 5)
print(value)
type(value)
# Nothing returned? Of course, we didn't return any value from the function
# NOTE!! Python has its own data type for "nothing"; for when, for example, a function returns "nothing"
```

```python
# If we want a function to return a value, we use the return statement
def return_substract(num_1, num_2):
    return num_1 - num_2
```

```python
return_substract(10, 5)
```

```python
# Here, we want to find the largest element in a given list
def find_max_in_list(some_list):
    max_element = some_list[0]
    length = len(some_list)
    for i in range(1, length):
        if some_list[i] > max_element:
            max_element = some_list[i] # max_element gets updated whenever we find a larger number
    return max_element # Until we eventually return the largest number we found
```

```python
num_list = [5, 4, 6, 7, 2, 10, 11, 3]
# Thus, after all the iterating over the elements, we've returned the largest element
find_max_in_list(num_list)
```

```python
# We can also store these values from a function into a variable
num_list2 = [45, 15, 63, 57, 82, 12]
highest_num = find_max_in_list(num_list2)
print(highest_num)
```

### Multi-return Statements

```python
# Within a function, it is possible to return multiple values at once
def add_and_sub(x, y):
    add_result = x + y
    sub_result = x - y
    return add_result, sub_result
add_and_sub(25, 17)
# Note the results are returned as a tuple?
```

```python
# But how do we receive and/or store these values? Recall how we could destructure values from tuples?
list1 = (2, 4)
a, b = list1
print(a, b)
```

```python
# We can do the same thing with the values returned, storing them in variables like so
result_1, result_2 = add_and_sub(25, 17)
print(result_1)
print(result_2)
```

```python
# Likewise recall, if we don't care or need a value from destructuring, we can use placeholders
a, b, _, c, _ = [5, 10, 15, 20, 15]
print(a, b, c)
```

```python
# Here, we ignore the sub_result; we just want the add_result here
result_1, _ = add_and_sub(40, 33)
print(result_1)
```

```python
# And the reverse
_, result_2 = add_and_sub(40, 33)
print(result_2)
```

```python
# Not only can we have multiple values returned, but we can have multiple return statements
def positive_or_negative(num):
    if num > 0:
        return "positive"
    elif num == 0:
        return "zero"
    else:
        return "negative"
```

```python
# Same function, three different arguments, three different values returned
print(positive_or_negative(25))
print(positive_or_negative(0))
print(positive_or_negative(-2))
```

## Keyword Arguments

### Invoking Functions

```python
# Issues with positional arguments arise if you have a lot of arguments, and mixing them around is easy and can lead to incorrect returns
```

```python
def total_score(math, physics, chemistry, biology):
    print("Math:", math, "Physics:", physics, "Chemistry:", chemistry, "Biology:", biology)
    return math + physics + chemistry + biology
```

```python
total_score(100, 78, 63, 88) # Which of these scores belong to what subject? We can't tell from here alone; it's easy to make a mistake
```

```python
# Python solves this by offering support for keyword arguments, where we can give, for example, (math = 25) in the arguments
```

```python
# We'll doubly show this, by having our keyword arguments in different positions than they're defined in the arguments
total_score(biology = 88, math = 100, chemistry = 63, physics = 78)
# We see that Python figures out what needs to go where, by us providing what argument name we're providing the values for
```

### Nuances

```python
# But what if only some of these are keyword arguments?
total_score(biology = 88, 100, 63, physics = 78)
# Here we see that positional arguments *must* be before keyword arguments
```

```python
# What about this now then?
total_score(100, 63, biology = 88, physics = 78)
# No, still an error; this is because python assumes the 2nd positional argument will be for the second argument; physics
# But we already defined it as a keyword argument; Python is not smart enough to realise that, and so throws an error
```

```python
# Thus, we must make sure out keyword arguments only refer to arguments our positional arguments don't touch/reach; here, the correct way
total_score(100, 78, biology = 88, chemistry = 63)
```

## Default Arguments

```python
# One issue with arguments is that if we forget the input for even one of these arguments, our function will return an error
def our_function(arg1, arg2, arg3):
    print(arg1, arg2, arg3)
```

```python
our_function("too little", "arguments")
```

```python
# To address this issue, we can use default arguments, where it assumes a certain value if we don't give an argument
# For example, recall how the sorted function has a keyword argument 'reverse'; it has a default argument for reverse as 'reverse=False'
```

```python
# Here we define a function where, by default, enrolled = True
def function_with_defaults(name, school, enrolled=True):
    print("Name:", name)
    print("School:", school, ", Enrolled:", enrolled)
```

```python
# Notice that, without supplying 3 arguments, it hasn't returned an error; because we've provided a value for lack of an argument; a default
function_with_defaults("Sally", "Highvale")
```

```python
function_with_defaults("Sally", "Highvale", True)
```

```python
# We have to explicitly say enrolled is False to override the default argument
function_with_defaults("Sally", "Highvale", False)
```

```python
# But of course, if we still leave out an argument without a default, we'll still return an error
function_with_defaults("Sally", enrolled=False)
```

```python
# You can define as many default arguments as you want; they can all have default values if you want
def function_with_all_defaults(name = "Tom", school="Highvale", enrolled=True):
    print("Name:", name)
    print("School:", school, ", Enrolled:", enrolled)
```

```python
function_with_all_defaults()
```

```python
function_with_all_defaults(enrolled=False)
```

## Variable Length Arguments

### *args Variable (Positional Variable-Length Arguments)

```python
# Aside from default arguments, it's also possible to set up functions that can take a variable amount of arguments
```

```python
# Taking another built-in function as an example, we know the print function can take no, one, or multiple arguments at once
print()
print("Red")
print("Red", "Green", "Blue")
```

```python
# Lets do some attempts to mimic the behaviour
```

```python
# This will only accept the arguments it's asking for, no more, no less, so this isn't correct
def print_fn(string_1):
    print(string_1)
print_fn()
```

```python
# If we include a default argument, we've managed some of the behaviour of variable length arguments, but not fully
def print_fn(string_1 = "red"):
    print(string_1)
print_fn()
```

```python
# But it still doesn't work with more arguments than it's asking for
print_fn("Green", "Blue")
```

```python
# Here then is how we have a function accept a variable amount of arguments
# NOTE!! It doesn't have to be called 'args', just so long as it has the asterisk '*' in front of it
def print_fn(*args):
    args_type = type(args)
    print(args_type)
    print(args)
```

```python
print_fn()
# Here we see that *args is representive of a tuple; it's how Python stores the variable amount of arguments
```

```python
print_fn("Greg", "Jane", "Harry", "Ron", "Jenny")
# No matter how many arguments, they can all fit in the tuple
```

```python
# When we pass in arguments such as a list, Python doesn't automatically spread the list across a tuple; it is a single element within the tuple
names_list = ["Greg", "Jane", "Harry", "Ron", "Jenny"]
print_fn(names_list)
```

```python
# If we want to spread it, we use *argument to spread the list to be put into the tuple
print_fn(*names_list)
```

### Combinations

```python
# We can combine the use of positional arguments and variable length arguments together to create powerful functionality
# Note our usage of two positional arguments, school and city, and our variable length argument, students
def students_in_school(school, city, *students):
    print("School:",school)
    print("City:",city)
    print("Students:",students)
```

```python
students_in_school()
# Note the error doesn't mention "3" arguments required? We only have 2 positional arguments
```

```python
# This function will need us to provide at least 2 arguments for the positional arguments
# but after that we can provide as many or as little students as we want
students_in_school("Highvale", "Melbourne")
```

```python
names = ["Julie", "Craig", "Samuel", "Sebastian"]
students_in_school("Highvale", "Melbourne", *names)
```

```python
students_in_school("Highvale", "Melbourne", "Julie", "Craig", "Samuel", "Sebastian")
```

```python
# We know that positional arguments must come first when using keyword arguments in functions
students_in_school(school="Highvale", "Kansas")
```

```python
# But there are also limitations with variable length arguments
# Say we define the variable length argument first in the order
def students_in_school(*students, school, city):
    print("School:",school)
    print("City:",city)
    print("Students:",students)
```

```python
students_in_school("Benjamin", "Harvard", "Hollywood")
# This will return an error, because Python believes *all* the arguments belong to the variable length argument; students
```

```python
# If we defined a function like this, the only way to correctly pass in arguments for school and city are to use keyword arguments
# NOTE!! Essentially, if your first argument is a variable length argument, all those after must be keyword-only arguments
students_in_school("Benjamin", school="Harvard", city="Hollywood")
```

### **kwargs Keyword (Keyword Variable-Length Arguments)

```python
# Python provides another way to specify variable-length arguments as an input for a function; keyword arguments (kwargs)
# When using double asterisk, Python will pack the kwargs into a dictionary, as opposed to a tuple with single asterisk
# In thi dictionary, the the keys are the variables names and the values are the variable values 
```

```python
def student_details(**kwargs):
    print(type(kwargs))
    print(kwargs)
```

```python
# kwargs also allows no arguments
student_details()
```

```python
student_details(name="John")
# Note the dictionary key:value pair
```

```python
student_details(name="John", age=18)
```

```python
# You don't have to call this kwargs, but it *is* a common name for such an argument; for example, here it's called details
def student_details(**details):
    for key, value in details.items():
        print(key, value)
```

```python
student_details(name="John", age=18, college="Boston University")
```

```python
# kwargs (again, standing for keyword arguments), means Python is expecting keyword arguments; if we provide positional, it will not work
student_details("John", 18, "Boston University")
```

```python
# Much like we passed in a list to *args, we can pass in a dictionary to **kwargs
details_dict = {"name": "John", "age": 18, "college": "Boston University"}
```

```python
# However, also like *args, it won't accept the list natively
student_details(details_dict)
```

```python
# We need to tell Python to unpack the details dictionary, by passing it in with '**'
student_details(**details_dict)
```

```python
# Here is an example where we check if a key exists, only printing the value if True
def student_details( ** details):
    if 'name' in details:
        print("Name:", details['name'])
    if 'age' in details:
        print("Age:", details['age'])
    if 'school' in details:
        print("School:", details['school'])
```

```python
student_details(age=18)
```

```python
student_details(school="Uni", age=18, name="John")
```

### Using Multiple Types of Variable-Length Arguments 

```python
# Here is an example where we're using both positional variable-length arguments and keyword variable-length arguments
# student_names will be packed up into a tuple as a positional, and college_details will be packed into a dictionary as a keyword
def students_in_college(*student_names, ** college_details):
    print ("Students--")
    for s in student_names:
        print(s)

    print()

    print("College Details--")
    for key, value in college_details.items() :
        print(key, value)
```

```python
students_in_college("Allison", "Bob", "Charlie", name="Stanford", city="Palo Alto")
```
