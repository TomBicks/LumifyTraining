# Conditional Statements & Loops: While Loops in Python

## An Introduction to While Loops

```python
# A while loop will continue to execute until its condition is true
number = 20
```

```python
while number < 20:
    print(number)
```

```python
# This code will execute infinitely; while loops in python won't detect if its infinite and stop; you must be *very* careful with while loops
#while number == 20:
    #print(number)
```

```python
# Here, we have code inside the while loop that will eventually make the coniditon false
number = 20
while number < 25:
    print(number)
    number += 1
# Eventually the number will reach 25 and the condition number < 25 will return false, stopping the while loop
```

```python
# The condition can have arithmatic included; it can be variables, or even calculations
number = 20
while number < (25+5):
    print(number)
    number += 1
```

## Basic While Loops

```python
# Here are some examples of while loops in action
```

```python
name = "Thomas"
while name == "Thomas":
    print("%s is the current name" % name)
    name = "Kingsley"
print("while loop has stopped")
```

```python
# Here is a more real-world example, with a password checker
user_password = ""
while user_password != "secret":
    user_password = input("Enter password:")
print("Access Granted.")
```

```python
num_value = int(input("How many times would you like the name displayed?"))
name = input("Which name would you like to display?")
count = 1

while count <= num_value:
    print(count, ":", name)
    count += 1
```

```python
# Using range, we check whether num is in the range between 0 (inclusive) and 10 (exclusive)
num = 0
while num in range(0, 10):
    print(num)
    num += 1
# This is where for and while loops differ greatly; for loops incremenet intrinsicaly; while loops, we need to introduce that functionality manually
```

```python
# Here is a loop which counts the number of digits in a number
value_int = int(input("enter an integer: "))
length = 0

# This loop will divide the number by 10, until it is less than 0, meaning we have divided the number into the decimals which, as an int, are removed
while value_int > 0:
    value_int //= 10 # This is the shorthand for division and equals, like +=, -=, or *=
    length += 1
print("number of digits:" , length)
print("value_int at end:", value_int)
```

```python
# While loops can also be nested within another while loop; take this example, which prints all pairs of x and y values
x = 1
while x < 5:
    y = 1
    while y < 5:
        print(x,y)
        y += 1
    x += 1
```

## Single-line While Loops

```python
# Here, we show how while loops can be condensed into a single line
# For reference, it's not actually any smaller, we've just used ';' to specify where the line breaks would've been
x, y = 0, 8
while(x<y): x += 1; print(x)
```

```python
# Thus, this is identical to this; the only difference is the use of indentation and new lines versus the usage of ';' on one line
x, y = 0, 8
while(x<y): 
    x += 1
    print(x)
```

## Evaluating Complex Data with While Loops

```python
# Printing each element of a list, using len(list) to make sure the list specifies the range of iterations
num_list = [20, 30, 40, 50, 60]
count = 0

while count < len(num_list):
    print(num_list[count])
    count +=1
print("There are",count,"elements in this list")
```

```python
total = 0
count = 0
while count < len(num_list):
    total += num_list[count]
    print(total)
    count += 1
print("Sum of the list is",total)
```

```python
# It should be noted however that these tasks, where we intend to iterate over the list a specified amount of times, is better handled by for loops
# Notice how this for loop requires less variables to handle the same task and result
# We also don't have to worry as much about our index going out of bounds in a for loop
total = 0
for element in num_list:
    total += element
    print(total)
print("Sum of the list is",total)
```

```python
# Here is an example where we've been careless, and the count will eventually go out of our lists index range
names = ["Tom", "Sally", "Stephen", "Amanda", "Greg"]
```

```python
count = 0
while count < 7:
    print("Count is at", count)
    print(names[count])
    count += 1
# Notice how the 'index' we set up, without being careful, has easily gone outside the index range of our list?
```

```python
# Again, this is much more a job for a for loop; this doesn't mean while loops aren't powerful, but they should be used only when needed
for name in names:
    print(name)
```

```python
# However, some calculations are equally easy or difficult for both for loops and while loops
# Here is an example where we're appending values to two separate lists
even_list = []
odd_list = []

num = 0
while num <= 20:
    if num % 2 == 0:
        even_list.append(num)
    else:
        odd_list.append(num)
    num += 1
print(even_list)
print(odd_list)
```

```python
even_list = []
odd_list = []

# Remember, we want 20, so range(21) is what we use
for num in range(21):
    if num % 2 == 0:
        even_list.append(num)
    else:
        odd_list.append(num)
print(even_list)
print(odd_list)
```

```python
# While loops end up often easier if we want to use an index for multiple lists
# For example, here we want to access and pair the values from the name and age lists together; we use index to access both values at the same index
# NOTE!! We assume here that both lists are of the same length and can be paired
name_list = ["Tom", "Jerry", "Matthew", "Jeremy", "Susan"]
age_list = [25, 52, 37, 28, 33]

index = 0
while index < len(name_list):
    print(name_list[index], "is", age_list[index])
    index += 1
```

## Exit a While Loop using the break Statement

```python
num = 10
while num in range(10, 100):
    print(num)

    # We'll usually have break statements behind conditionals, such as an if statement
    if num == 50:
        print("Num is 50; breaking while loop")
        break
        
    num += 10
```

```python
haystack = ['hay', 'hay', 'hay', 'hay', 'hay', 'needle', 'hay', 'hay']

i = 0
while i < len(haystack):
    if haystack[i] == "needle":
        print("Found needle!")
        break
    i += 1
# The else block will run after the while loop ends naturally; break won't run the else block
else:
    print("Needle wasn't found")
```
