## Iterating over Elements in a List

```python
# In a for loop, we specify what we want the value of the iteration to be called (here, we use letter)
for letter in ['H', 'e', 'l', 'l', 'o']:
    print(letter)
```

```python
# As we know, strings can be thought of as lists of characters; thus, we can iterate over a string the same way
for letter in 'Hello':
    print(letter)
```

```python
# The string 10 does not convert to a number, but is iterated over like a regular string
for digit in '10':
    print(digit)
```

```python
# numbers are not iterable, and thus cannot be used here
for digit in 10:
    print(digit)
```

```python
# Say we have a list with mixed data; the data types are not consistent across it
mixed_list = [145, 10.5, 1+3j,
              True, "Python",
              (0,1), [2,-5],
              {"Class":"V", "Section":"A"}]
```

```python
# As we can see, we *can* iterate over such a list, even with all the data types
for item in mixed_list:
    print ("Type of ", item, " is ", type(item))
```

## Iterating over Elements in a Dictionary

```python
student_scores = {'John': 80,
                  'Sam': 60,
                  'Jill': 50,
                  'Bob': 96}
student_scores
```

```python
# If we iterate over a dictionary, what will it return? Key? Value?
for student in student_scores:
    print(student)
# Keys it is!
```

```python
# This is also identical to iterating over .keys()
for key in student_scores.keys():
    print(key)
```

```python
# To grab the scores, we need to specify for the values (as we have in the past)
for scores in student_scores.values():
    print(scores)
```

```python
# But we can still grab both the keys and values at once, by slotting the key back into the dictionary to return the corresponding value
for key in student_scores:
    print("Key value pair = ", key, ":", student_scores[key])
```

```python
# However, there is an even simpler and more straightforward way of grabbing these pairs; the .items() function
for key_value_pair in student_scores.items():
    print("Key value pair = ", key_value_pair)
# As you can see, each item is returned as a tuple
```

```python
# We can destructure the key:value pair returned by .items() like this
for student, score in student_scores.items():
    print("Student =", student, ", Score =", score)
```

## Iterating over Elements in a Set

```python
# As we've already established, sets have no inherent order; therefore, you cannot trust them to be iterated over in a consistent manner
my_set = {'a', 'b', 'c', 'd'}
```

```python
for item in my_set:
    print(item)
# As you can see, the order the items are printed in is completely different to the "order" we created them in when we initialised the set
```

## The Else Block of a For Loop

```python
# Python's For loops have an additional feature uncommon amongst programming languages; an else statement
numbers = [2, 3, 5, 7]
my_string = input("Enter a string:")
```

```python
# An else statement of a for loop will run when the for loop reaches the end of its iterations with no errors
# So if a for loop returns an error, or if it exits prematurely, the else block will not run
for num in numbers:
    print(num)
else:
    print("No more numbers left")
```

```python
for char in my_string:
    print(char)
else:
    print("No more characters left")
```

## Nested Control Structures in a For Loop

```python
# We can use If statements and else blocks within for loops
cities = ['New York', 'Kansas', 'Dallas']
```

```python
for city in cities:
    if city == 'New York':
        print(city, "= New York!")
    else:
        print(city, "!= New York")
```

```python
# For loops can also be nested within each other
names = [['Tom', 'Sam', 'Dave'],
        ['Brett', 'Linda', 'Steve']]
```

```python
for sublist in names:
    for name in sublist:
        print(name)
```

```python
# We can also show this off with separate lists; printing a colour and object combination for each colour
colours_list = ['Red', 'Green', 'Blue']
objects_list = ['Pen', 'Marker', 'Crayon']
```

```python
for colour in colours_list:
    for obj in objects_list:
        print(colour, obj)
# See that each of the 9 possible combinations is printed out
```

## An Introduction to the range Function

```python
# By default, the number of iterations is equal to the size of the structure; n elements will be iterated over n times
# The range function can explicitly define the number of iterations over a structure
my_range = range(5)
my_range
# As we can see, what a range object actually is; a sequence of numbers from x to y
```

```python
# We can see this more clearly by converting the range object into a list
list(my_range)
# here, the range is from 0 to 4; remember, index starts at 0, so if we wanted to iterate over 5 elements, the last index would be 5 - 1 = 4
```

```python
# Range also works with negatives numbers
list(range(-5, 5))
```

```python
# We can also show this off in a for loop itself
for i in range(6):
    print(i)
```

```python
# If we don't need the variable from iteration (e.g. 'i') then we can replace it with '_' (placeholder)
for _ in range(3):
    print("Hello!")
```

## Setting Intervals in a Range

```python
# Mathematical functions can be passed into range
# Reminder that this means 2 to the power of 2
# Also recall that the final interval value will be 1 less than the range
for i in range(2 ** 2):
    print(i ** 2)
```

```python
# We can reverse the range too, with the reversed function
x = range(5)
for i in reversed(x):
    print(i)
```

```python
# If you recall how we specified start and end index and step for slice operations, this will be very similar
range_specified = range(5, 10, 1) # Start index, end index, step interval
list(range_specified)
```

```python
range_specified = range(5, 10, 2) # Start index, end index, step interval
list(range_specified)
```

```python
range_specified = range(5, -10, -2) # Start index, end index, step interval
tuple(range_specified)
```

```python
# Remember also that the end index is not inclusive, just like slice functions!
range_specified = range(-2, -10, -2) # Start index, end index, step interval
list(range_specified)
# No -10 in sight
```

```python
# We can also pass these 'start', 'stop' and 'step' values in as arguments
start = 2
stop = 10
step = 2
for number in range(start, stop, step):
    print(number)
```
