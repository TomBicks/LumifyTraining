## Formating Operations with Strings in Python

```python
# We can perform string concatenating manually, but it can be hard to read
dog_sound = 'barks'
print('My dog ' + dog_sound + '!')
```

```python
# We can instead, using 'f' and '{}', interpret anything inside the curly braces within the string as an expression
print(f'My dog {dog_sound}!')
```

```python
# Without the f, it's all interpreted as a single string
print('My dog {dog_sound}!')
```

```python
# We went over special characters before, needing backslash as an escape character to work
# This one includes a tab in the string
print("John asked \t'How are you?'")
```

```python
# But how do we include in backslash in a string as part of the text? Another backslash!
print("C:\\Users\\Me")
```
