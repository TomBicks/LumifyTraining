## Introducing the break Statement

```python
# If we want to terminate a loop partway through, for example if we've found a value we are looking for, we can use the break statement
for letter in 'python':
    if letter == "o":
        break
    print(letter)
# Notice the o is never printed? Whilst we encountered 'o', we broke out of the loop before it could be printed
```

```python
# This even more clearly shows it is the break statement that is exiting the loop
for letter in 'python':
    print(letter)
    if letter == "o":
        break
```

```python
# As we've gone over before, if we break prematurely from a loop, the else block of a for loop is *not* run
for char in 'icecream':
    if char == 'c':
        print('The letter %s was found' % char)
        break
else:
    print("The letter does not exist in the string")
```

```python
# And here we see, without anything to satisfy the if statement leading to the break statement, the else block is ran
for char in 'python':
    if char == 'c':
        print('The letter %s was found' % char)
        break
else:
    print("The letter does not exist in the string")
```

## The continue Statement

```python
# What if we want to stop a single iteration of a for loop, but don't want to break out of it entirely? We use the continue statement
for letter in 'string':
    if letter == "i":
        continue
    print(letter)
print("The end")
# Here, we see only the iteration where letter = 'i' is stopped; we 'continued' to the next iteration, not invoking the print function
```

```python
for letter in 'Students':
    if letter == "e" or letter == "s":
        continue
    else:
        print(letter)
```

```python
num_days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec" ]
new_dict = dict()
```

```python
# Remember that the zip function "zippers" two lists or tuples together into tuple pairs of sorts
for i, j in zip(months, num_days):
    new_dict.update({i:j})
    if new_dict[i] < 29:
        continue
    else:
        print(i, "month has", new_dict[i], "days")
```

## The pass Statement

```python
# The pass statement is used to denote the fact that nothing needs to be done
# The pass statement does nothing. It can be used when a statement is required syntactically but the program requires no action, such as when we need a placeholder
for letter in 'string':
    if letter == 'r':
        pass
    print(letter)
```

```python
my_str = input ("Input a string: ")

digits = 0
letters = 0

for i in my_str:
    if i.isdigit():
        digits = digits + 1
    elif i.isalpha():
        letters = letters + 1
    else:
        pass # Here we can use it as a placeholder, but the code works just as well without the else block
print("Digits = ", digits)
print("Letters = ", letters)
```

## Introducing Comprehensions

```python
# List Comprehensions allows us to create lists from other lists, using for loops, in very brief code
num = range(10)
[x * x for x in num]
# Here we have *very* quickly and concisely created a list of elements where each element is the index cubed
```

```python
# Here we produce a list of letters from the word "anxiety", in one line
letters = [letter for letter in "anxiety"]
# For a more detailed explanation, we are using the value of the iteration, letter, to determine the element, before the 'for'
# Above the element was 'x * x' (e.g. x (iteration) was 3, the value of the element than was 3 * 3 = 9
# Here, the element value is just the letter; the value of the iteration
letters
```

```python
list_of_words = ["this", "is", "a", "list", "of", "words"]
# We can transform the resulting element here too
items = [word[0].upper() for word in list_of_words] # The element we're producing is the first character of each word, then made uppercase
print (items)
# Witness the power of list comprehensions; all this in one line!
```

```python
# The element returned doesn't have to be just a string or int either; here we return a tuple of values
stationery = ["Pen", "Marker", "Ink"]
colors = ["Red", "Blue", "Green"]

combined = [(i, j) for j in stationery for i in colors ]
combined
```

```python
# It should be noted that, whilst we've only created lists here, changing the outer brackets allows us to create sets, tuples, and even dictionaries
letters = {letter for letter in "anxiety"}
print(letters)
```

## TODO!! Applying Conditions in Comprehensions

```python
# We can even include if statements in list comprehensions
# Here, we base it off our first example in our previous topic, but instead we only want the odd results (numbers which aren't divisible by 2)
num = range(10)
[x * x for x in num if x % 2 != 0]
# See how we've managed to functionally filter out the even results?
```

```python
# Here is another example using if statements to extract the vowels from a string, by checking if the letter can be found in our list of vowels
[i for i in "mathematics" if i in ['a', 'e', 'i', 'o', 'u']]
```

```python
# In our third example with if statements, we pull any number characters from a string, then convert them into real integers
my_string = "Hello 12345 World6!"
numbers = [int(x) for x in my_string if x.isdigit()]
numbers
```

```python
# Ditto but with letters
letters = [x for x in my_string if x.isalpha()]
letters
```

```python
# Nested if statements are also possible, though function the same as if including the 'and' keyword
number_list_nested = [number for number in range(51) if number % 2 == 0 if number % 5 == 0]
number_list_and = [number for number in range(51) if number % 2 == 0 and number % 5 == 0]
print(number_list_nested)
print(number_list_and)
```

```python
# If statements don't just have to be aassociated with the for loop; they can also be used in regards to the element returned
# Here, for each iteration, the value returned is "Even" if the index is even, and "Odd" if the index is odd
num = ["Even" if i % 2 == 0 else "Odd" for i in range(10)]
num
```
