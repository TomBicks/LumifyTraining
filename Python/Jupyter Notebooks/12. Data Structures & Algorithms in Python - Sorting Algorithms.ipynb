Here is a list of a few potential interview questions, with some explanations on what these mean underneath each:

- Is the sort adaptive?
    - Does the algorithm work better if the input array is already sorted to soem degree.
    - Adaptive sorts are able to adapt themselves to the input, and will often break early if the input is already sorted.

You can see here that it does help to have an in-depth understanding of the specific sorting technique which you are implementing, and it is important for you to know the various trade-off in the different sorting algorithms.

## The Selection Sort

*Detailed instructions, descriptions, and implementations of Selection Sort can be found at [W3Schools](https://www.w3schools.com/dsa/dsa_algo_selectionsort.php)*

*Details of the advantages and disadvantages of Selection Sort can be found at [YouAcademy](https://youcademy.org/selection-sort-pros-cons/), including when to use and when to avoid using Selection Sort*

***A summary of the advantages and disadvantages of Selection Sort are:***
- **Advantages**
    - Simple Implementation
    - In-Place Algorithm
    - Good for Minimising Writes
- **Disadvantages**
    - Poor Time Complexity
    - Inefficient for Large Datasets
    - Not Adaptive
    - Unstable Sorting
- **Key Takeaways**
    - Selection sort is **simple** to understand and implement, with minimal memory usage (O(1) space complexity).
    - Its main drawback is the O(n^2) time complexity, making it inefficient for larger datasets.
    - It’s best suited for **small datasets** or situations where **minimizing writes to memory** is a priority.
    - It’s **not adaptive** to partially sorted input.
    - It’s **not stable** by default (though stable variants exist).
    - The algorithm performs at most n-1 swaps, which can be beneficial in specific scenarios.

The selection sort takes an array. At each iteration, we will select one element, then compare it to every other element in the list, in order to find the smallest element. For example, in our first iteration, we select the first element, then find the smallest number by comparison, then swap the elements, so the smallest is now our first element.

The process is then repeated to find the second smallest element, excluding the first element set in place, swapping them until all elements are sorted.

***The basic steps of the Selection Sort algorithm are:***
1. Go through the array to find the lowest value.
2. Move the lowest value to the front of the unsorted part of the array.
3. Go through the array again as many times as there are values in the array.

Each iteration, the number of elements we need to compare to find the smallest element is reduced by one, as we set our smallest elements in place.

Due to the process of going through each element and "selecting" our next smallest value to sort, this is why it is called a *selection sort*.

The Selection sort algorithm, for each element, will go through the array to find the smallest value for that index. So, for an array of n values, there will be n iterations over the outer loop. With each iteration of the outer loop, the inner loop will make comparisons with the selected element against every other element to try and find the smallest value for that index; the inner loop will reduce the amount of elements it needs to iterate over by 1 after each iteration of the outer loop, as we set smallest values in place and no longer need to compare them. Thus, ***the time complexity of Selection Sort is O(n^2)***.

As we've mentioned, the number of comparisons Selection Sort needs to make lessens after each iteration of the outer loop, as we set smallest values in place; so (n-1) for the first iteration, (n-2) for the second, etc. Whilst this does roughly equate to n(n-1)/2 or (n^2 - n)/2, we know that Big O notation disregards constants and lower-order terms (lower-order referring to how n is order 1, and n^2 is order 2, so we ignore n), meaning ***the number of comparisons Selection Sort makes is O(n^2)***. However, because Selection Sort only makes n swaps overall, once per each iteration of the outer loop, ***Selection Sort performs O(n) swaps***.

Selection Sort only use one or more variables to store the index of the current minimum value. Likewise, ***it is an "in-place" sorting algorithm***, meaning the array it is given to sort is itself modified and sorted, it does not need extra space for an output array. These together mean it takes up a constant amount of space, not relating to the size of n, and thus ***Selection Sort is said to take up O(1) extra space***.

Selection Sort may result in elements which are of equal value ending up sorted differently after the algorithm is performed, due to the possibility that some elements can be moved from the beginning of the array to the end. As a result, ***it is an unstable sort***. It should be noted that is *is possible* to make a Selection Sort stable, but it often adds further complexity to the algorithm. 

### Version 1 - Correct, but not the best

```python
def selection_sort(input_list):
    print("--- Selection Sort ---")
    n = len(input_list)
    num_iterations = 0

    print(f"List at start is {input_list}")

    # We can check for 0 to n-1 because, with how selection sort works, the last element should already be the largest
    for i in range(n-1):
        min_index = i

        # Checks for i+1, because after each iteration of the outer loop, i number of smallest elements have been set in place
        # We don't need to iterate over those and check them; we already know they're the smallest for that index
        for j in range(i+1, n):
            num_iterations += 1
            if input_list[j] < input_list[min_index]:
                print(f"Current element {input_list[j]} is smaller than {input_list[min_index]}. Setting new smallest element for index {i}. ")
                min_index = j

        # We then pop the smallest value from the list and add it in as the next smallest element
        min_value = input_list.pop(min_index)
        input_list.insert(i, min_value)
        print(f"Popping {min_value} and adding to to index {i}")
        print(f"List is currently {input_list}")

    print(f"List sorted after {num_iterations} iterations!")
    print(input_list)
```

```python
input_list = [15, 32, 26, 11, 36, 19, 42, 44, 14]

selection_sort(input_list)
```

Whils this is indeed a selection sort, it was not done as well as it could have been. 

Notice the usage of pop and insert? When we pop an element, every element after it needs to shuffle to the left once to fill in the gaps. Likewise, when we insert an element, ever element after it needs to shuffle to the right one. Whilst in small lists this isn't really an issue, again we think of if the inputs were *huge* when dealing with complexity. If we had a huge list as an input, these shifting operations would take a lot of processing.

### Version 2 - More efficient than Version 1

Our best bet with a selection sort is to instead swap the smallest element with the current iterations element.

First iteration? We search the whole list and swap the smallest element with the first element. Second iteration? We search the whole list past the first element, then swap the next smallest element with the second element. etc.

```python
def selection_sort_better(input_list):
    print("--- Selection Sort (Better) ---")
    n = len(input_list)
    num_iterations = 0

    print(f"List at start is {input_list}")

    # We can check for 0 to n-1 because, with how selection sort works, the last element should already be the largest
    for i in range(n-1):
        min_index = i

        # Checks for i+1, because after each iteration of the outer loop, i number of smallest elements have been set in place
        # We don't need to iterate over those and check them; we already know they're the smallest for that index
        for j in range(i+1, n):
            num_iterations += 1
            if input_list[j] < input_list[min_index]:
                print(f"Current element {input_list[j]} is smaller than {input_list[min_index]}. Setting new smallest element for index {i}. ")
                min_index = j

        # We then swap the elements around, by passing in the values of the other element to each other
        print(f"Swapping {input_list[i]} and {input_list[min_index]}")
        input_list[i], input_list[min_index] = input_list[min_index], input_list[i]
        print(f"List is currently {input_list}")

    print(f"List sorted after {num_iterations} iterations!")
    print(input_list)
```

```python
input_list = [15, 32, 26, 11, 36, 19, 42, 44, 14]

selection_sort_better(input_list)
```

Now we note that, whilst the number of iterations remained the same, the sort (theoretically) ran much more efficiently. Whilst this efficiency can't be represented under Big O notation, it does show that there are additional challenges to writing functions beyond Big O notation.

## The Bubble Sort

*Detailed instructions, descriptions, and implementations of Bubble Sort can be found at [W3Schools](https://www.w3schools.com/dsa/dsa_algo_bubblesort.php)*

*Details of the advantages and disadvantages of Bubble Sort can be found at [YouAcademy](https://youcademy.org/bubble-sort-advantages-disadvantages/), including when to use and when to avoid using Bubble Sort*

***A summary of the advantages and disadvantages of Bubble Sort are:***
- **Advantages**
    - Simple Implementation
    - Easy to Detect Sorted Array
    - In-Place Algorithm
    - Stable Sorting
- **Disadvantages**
    - Poor Time Complexity
    - Inefficient for Large Datasets
    - Suboptimal Performance
    - Many Element Swaps
- **Key Takeaways**
    - Bubble Sort is **simple and memory-efficient** with O(1) space complexity.
    - Its main drawback is the O(n^2) time complexity for average and worst cases.
    - The best case is when the array is already sorted, resulting in O(n) time complexity
    - It’s best suited for **educational purposes** or **very small datasets**.
    - It is not suitable for practical use cases or large datasets due to its inefficiency.

Rather than finding the smallest element in every single iteration, bubble sort performs a series of pairwise comparisons. The effect is largely the same however, making this sort fairly similar to selection sort.

The major difference comes with the pairwise comparisons. Rather than only swapping the smallest element into place, as selection sort did, bubble sort's swaps at each inner iteration between a larger and smaller value mean that, as bubble sort continues, we may find that the list actually begins to be sorted even aside from the largest values we're sorting to the end of the list.

***The basic steps of the Bubble Sort algorithm are:***
1. Go through the array, one value at a time.
2. For each value, compare the value with the next value.
3. If the value is higher than the next one, swap the values so that the highest value comes last.
4. Go through the array as many times as there are values in the array.

With each iteration however, we are likely to find that the number of swaps that are required goes down, because the pairwise comparisons are, little by little, also naturally causing smaller elements to get sorted to the right, and larger elements sorted to the left.

In the worst case however, where you have an array which is sorted in reverse order (i.e. the worst possible array for a bubble sort), a swap will always be required at every single step.

The Bubble Sort algorithm loops through every value in the array, comparing it to the value next to it. So, for an array of n values, there must be n such comparisons in one inner loop. And after one inner loop, the array is looped through again by the outer loop, n times. ***This means the time complexity of Bubble Sort is O(n^2)***.

Likewise it will, for each of these n loops of n checks, in the worst case, make n^2 comparisons and n^2 swaps. Thus, ***Bubble Sort makes O(n^2) comparisons and O(n^2) swaps***. However, it ***can be made adaptive*** by breaking if a loop requires no swaps, which makes it slightly better on average,

Bubble Sort only uses a handful of variables at most to swap values, and as ***it is an "in-place" sorting algorithm***, meaning the array it is given to sort is itself modified and sorted, it does not need extra space for an output array. This means it takes up a constant amount of space, not relating to the size of n, and thus ***Bubble Sort is said to take up O(1) extra space***.

Bubble Sort does *not* swap two elements of equal value, only those that are smaller. For this reason, ***it is considered a stable sort***.

### Version 1 - Makes too many comparisons if

```python
def bubble_sort(input_list):
    print("--- Bubble Sort ---")
    n = len(input_list)
    num_iterations = 0
    comparisons = 0
    swaps = 0

    print(f"List at start is {input_list}")

    # We need to run the inner loop n-1 times, as it will already be sorted by then; n times would be a waste
    for i in range(n-1):
        print(f"Outer Loop {i}")
        # As we iterate, we set more and more of the largest numbers into place on the right
        # Hence, reduce inner iterations by i (0 to n-1)
        for j in range(n-1-i):
            num_iterations += 1
            print(f"Current element: {input_list[j]}, Next element: {input_list[j+1]}")
            comparisons += 1
            if input_list[j] > input_list[j+1]:
                print(f"Swapping {input_list[j]} and {input_list[j+1]}")
                input_list[j], input_list[j+1] = input_list[j+1], input_list[j]
                swaps += 1
                print(f"List is currently {input_list}")
    
    print(f"List sorted after {num_iterations} iterations, {comparisons} comparisons, and {swaps} swaps!")
    print(input_list)
```

```python
input_list = [15, 32, 26, 11, 36, 19, 42, 44, 14]

bubble_sort(input_list)
```

There is one slight issue that can occur with this function however. Imagine we have a list that's almost sorted. Our bubble sort has no way of detecting this, and will instead run through every inner and outer loop, making comparisons it doesn't need to make.

By checking if we made *no* swaps in the previous outer loop, we can improve this function, and prevent it from making comparisons it doesn't need to, in the next version.

### Version 2 - Making it adaptive

```python
def bubble_sort_better(input_list):
    print("--- Bubble Sort (Better) ---")
    n = len(input_list)
    num_iterations = 0
    comparisons = 0
    swaps = 0

    print(f"List at start is {input_list}")

    # We need to run the inner loop n-1 times, as it will already be sorted by then; n times would be a waste
    for i in range(n-1):
        print(f"Outer Loop {i}")

        # We hold whether any swaps have occurred this loop; False by default
        # This helps reduce the number of unrequired comparisons and swaps for already nearly completed lists
        swapped = False
        
        # As we iterate, we set more and more of the largest numbers into place on the right
        # Hence, reduce inner iterations by i (0 to n-1)
        for j in range(n-1-i):
            num_iterations += 1
            print(f"Current element: {input_list[j]}, Next element: {input_list[j+1]}")
            comparisons += 1
            if input_list[j] > input_list[j+1]:
                print(f"Swapping {input_list[j]} and {input_list[j+1]}")
                input_list[j], input_list[j+1] = input_list[j+1], input_list[j]
                swaps += 1
                swapped = True # We swapped values, so swapped is True
                print(f"List is currently {input_list}")
        # This will break the loop if we're already sorted, ensuring we don't spend an outer loop comparing all elements of a sorted list at the end
        if not swapped:
            print(f"No swaps occurred on Outer Loop {i}; list is sorted already!")
            break
    
    print(f"List sorted after {num_iterations} iterations, {comparisons} comparisons, and {swaps} swaps!")
    print(input_list)
```

```python
input_list = [15, 32, 26, 11, 36, 19, 42, 44, 14]

bubble_sort_better(input_list)
```

Now, whilst our improvement doesn't appear to have changed anything with our current list, lets take another example, where the list is already nearly solved.

```python
my_array = [7, 3, 9, 12, 11]
bubble_sort(my_array)
```

```python
my_array = [7, 3, 9, 12, 11]
bubble_sort_better(my_array)
```

We see that our second version of bubble sort, which cuts off our bubble sort outer loop if no swaps occurred in the last outer loop, has produced 3 less iterations and comparisons, because it never ran the last few outer loops; they were cut off

## The Insertion Sort

*Detailed instructions, descriptions, and implementations of Insertion Sort can be found at [W3Schools](https://www.w3schools.com/dsa/dsa_algo_insertionsort.php)*

*Details of the advantages and disadvantages of Insertion Sort can be found at [YouAcademy](https://youcademy.org/advantages-disadvantages-of-insertion-sort/), including when to use and when to avoid using Insertion Sort*

***A summary of the advantages and disadvantages of Insertion Sort are:***
- **Advantages**
    - Simple Implementation
    - Adaptive Algorithm
    - In-Place Algorithm
    - Stable Sorting
    - Efficient for Small Datasets
    - Ideal for Real-time Applications
- **Disadvantages**
    - Poor Time Complexity
    - Inefficient for Large Datasets
    - Many Element Shifts
    - Not Parallelizable
- **Key Takeaways**
    - Insertion Sort is **simple, stable, and memory-efficient** with O(1) space complexity.
    - Its main drawback is the **O(n²)** time complexity for average and worst cases.
    - It’s best suited for **small datasets** or **nearly sorted data**.
    - The algorithm performs well in practice for small arrays and can be faster than more complex algorithms in these cases.

Insertion Sort performs much like a card-player would when given a hand of random cards; they, one at a time, shuffle their cards into the correct order.

Insertion Sort goes across each element, before shuffling it to the left; the "sorted" part of the array. Then one by one, we move the elements across, comparing them to the rest of the "sorted" array to see where they need to go. For reference, the "sorted" part of the array is still the same array; we don't inherently use sub-arrays in this algorithm??? I THINK???

***The basic steps of the Insertion Sort algorithm are:***
1. Take the first value from the unsorted part of the array.
2. Move the value into the correct place in the sorted part of the array.
3. Go through the unsorted part of the array again as many times as there are values.

Assuming nothing is sorted, we compare the first two values. Assuming they are in order, we then compare the second and third value. Assuming the third element is smaller than the second, we swap the second and third. We then go back and compare this second element, previously the third, against the first element, to see if that too is larger than it. If not, we move on.
Then compare 3 and 4; if 4 is smaller, we then swap, then compare it against the 2nd element, then swap if its smaller, then the first if so, swapping if smaller.
Essentailly each iteration, the list should be sorted from the start onwards, growing with each iteration

Insertion Sort, given n values, must run the loop to insert/swap a value into its correct place approximately n times. It then, on average, must compare each value to about n/2 other values to find the correct place to insert/swap it. So whilst the complexity is technically n * n/2, we know under Big O notation we can discard constants and lower-order terms, so ***the time complexity of Insertion Sort is O(n^2)***.

Likewise it will, for each of these n loops of n checks, in the worst case, make n^2 comparisons and n^2 swaps. Thus, ***Insertion Sort makes O(n^2) comparisons and O(n^2) swaps***. However, it is very worth noting that ***Insertion Sort is adaptive***, meaning the more already sorted a list is, the closer complexity, comparisons and swaps draws to O(n) (if it's given an already sorted array, it will perform no swaps, running in O(n) time).

Insertion Sort only uses a handful of variables at most to retrain and swap values, and as ***it is an "in-place" sorting algorithm***, meaning the array it is given to sort is itself modified and sorted, it does not need extra space for an output array. This means it takes up a constant amount of space, not relating to the size of n, and thus ***Insertion Sort is said to take up O(1) extra space***.

Because Insertion Sort works left to right when selecting elements, and then right to left when inserting them, even if values are separated temporarily, their order will inevitably be preserved when they're placed alongside each other. For this reason, ***Insertion Sort is considered a stable sort***.

### Insertion Sort vs Bubble Sort

Adaptive sort algorithms perform very well when working with nearly sorted data. However, when compared to other adaptive sorts, namely Bubble Sort, Insertion Sort is seen to outperform it slightly.

This is largely due to this; **Bubble Sort makes far more comparisons and swaps than Insertion Sort**, with the difference only growing as n does. Bubble Sort needs to iterate back over an entire array to check its been completely sorted, only exiting when no swaps have been detected, whereas with an Insertion Sort, we can stop the comparisons once we find that an element can no longer move towards the left of the array.
Whilst when comparing worst-case, Bubble Sort and Insertion Sort seem on par, the average shows that Bubble Sort swaps elements far more than Insertion Sort. 

### Insertion Sorts usage in 'Divide-and-Conquer' Algorithms

Insertion Sort itself, due to its low overhead and simple operations, often outperforms sophisticated algorithms like Quick Sort or Merge Sort. However, where adaptive sorts struggle, including Insertion Sort, is against very large lists, where the elements can be extremely far away from where they belong (the worst of which is a list already sorted in reverse order).

To this end, this is why 'divide and conquer' algorithms, like the Shell Sort on the next topic, work so well with adaptive algorithms like Insertion Sort; they break the given array down into subarrays, far more manageable for the Insertion Sort to handle, working to Insertion Sort's strengths and helping mitigate its weaknesses.

### Version 1 - Too many shifts

```python
def insertion_sort(input_list):
    print("--- Insertion Sort ---")
    n = len(input_list)
    num_iterations = 0
    comparisons = 0
    swaps = 0

    print(f"List at start is {input_list}")

    # We assume the first value is "already sorted", so we start at index 1, and sort into the "sorted array" from there
    for i in range(1,n):
        print(f"--- Outer loop i = {i}")
        insert_index = i # By default, we'll be inserting the current element directly after the elements of the "sorted array"
                         # unless elements in the sorted array are bigger than the current element
        current_value = input_list.pop(i)
        print(f"Popping current value {current_value}")
        for j in range(i-1, -1, -1):
            print(f"------ Inner loop j = {j}")
            num_iterations += 1
            comparisons += 1
            print(f"Is {input_list[j]} greater than {current_value}? {input_list[j] > current_value}")
            if input_list[j] > current_value:
                # If we found a value in the sorted array that's larger than our current value, we know we have to work backwards to check
                # where out current value fits into the sorted array
                print(f"{input_list[j]} at index {j} is larger than current value {current_value}. Setting insert index to {j}")
                insert_index = j
        input_list.insert(insert_index, current_value)
        print(f"Inserting current value {current_value} at index {insert_index}")
        swaps += 1
        print(f"List is currently {input_list}")
    
    print(f"List sorted after {num_iterations} iterations, {comparisons} comparisons, and {swaps} swaps!")
    print(input_list)
```

```python
input_list = [15, 32, 26, 11, 36, 19, 42, 44, 14]

insertion_sort(input_list)
```

You might recall how we've said that whenever an element is removed or added from an array, shift operations have to occur, working behind the scenes to shift all the elements to fill in the gap or make room for an insertion.

Thus, whilst this example works, there is a lot of shift operations occuring, even when an element *doesn't need to be moved*!.

Not only this, but we can also see that even when we've found where our current element needs to be inserted, we're still iterating over the rest of the sorted array in reverse.

Improvements can be made by only shifting the values necessary, rather than the whole array, as well as breaking the inner loop when we've found where the current element needs to go. We'll show this in the next version.

### Version 2 - No Shifting Required and Loop Breaks when needed

```python
def insertion_sort_better(input_list):
    print("--- Insertion Sort (Better) ---")
    n = len(input_list)
    num_iterations = 0
    comparisons = 0
    swaps = 0

    print(f"List at start is {input_list}")

    # We assume the first value is "already sorted", so we start at index 1, and sort into the "sorted array" from there
    for i in range(1,n):
        print(f"--- Outer loop i = {i}")
        insert_index = i # By default, we'll be inserting the current element directly after the elements of the "sorted array"
                         # unless elements in the sorted array are bigger than the current element
        current_value = input_list[i]
        print(f"Current value is {current_value}")
        for j in range(i-1, -1, -1):
            print(f"------ Inner loop j = {j}")
            num_iterations += 1
            comparisons += 1
            print(f"Is {input_list[j]} greater than {current_value}? {input_list[j] > current_value}")
            if input_list[j] > current_value:
                # If we found a value in the sorted array that's larger than our current value, we know we have to work backwards to check
                # where out current value fits into the sorted array
                print(f"Setting {input_list[j+1]} at index {j+1} to {input_list[j]}")
                input_list[j+1] = input_list[j]
                print(input_list)
                print(f"{input_list[j]} at index {j} is larger than current value {current_value}. Setting insert index to {j}")
                insert_index = j
            else:
                # Now we break out of the inner loop when we've found where the current value needs to be set; saves on iterations/comparisons
                break
        print(f"Inserting current value {current_value} at index {insert_index}")
        input_list[insert_index] = current_value
        swaps += 1
        print(f"List is currently {input_list}")
    
    print(f"List sorted after {num_iterations} iterations, {comparisons} comparisons, and {swaps} swaps!")
    print(input_list)
```

```python
input_list = [15, 32, 26, 11, 36, 19, 42, 44, 14]

insertion_sort_better(input_list)
```

As we can see, when an inner loop founds where our current element is supposed to go; the first time, iterating backwards over the sorted array, it finds an element that is not bigger than the current element, the inner loop breaks, and we move on to the next outer loop.

Likewise, by swapping the values as we iterate backwards to find the right insert index (in this case actually overwriting the elements in reverse order, whilst keeping in memory what the element we overwrote was), we avoid needing to deal with shift operations entirely, saving on behind-the-scenes processessing.

Shell Sort is an improved version of Insertion Sort that works by comparing elements that are far apart instead of right next to each other. By starting with far-apart elements, it can move some out-of-place elements into the position faster than a simple nearest-neighbor exchange.

It does this by breaking up an array into several sub-arrays in a special way, then passing these arrays onto the Insertion Sort, returning sorted sub-arrays. It then rebreaks the array into slightly larger sub-arrays, sorting these. It then repeats this process until the entire array is sorted in one large use of Insertion Sort across the whole array, now largely having already been sorted, working towards Insertion Sort's strength as an adaptive sorting algorithm.

It is for this reason it is called a 'divide and conquer' algorithm, breaking up an array to better handle it in pieces.

Another divide-and-conquer algorithm, the Merge Sort algorithm sorts an array by first breaking it down into smaller arrays, and then building the array back together the correct way so that it is sorted

***The basic steps of the Merge Sort algorithm are:***
1. Divide the unsorted array into two sub-arrays, half the size of the original.
2. Continue to divide the sub-arrays as long as the current piece of the array has more than one element.
3. Merge two sub-arrays together by always putting the lowest value first.
4. Keep merging until there are no sub-arrays left.

The breaking down and building up of the array, to sort the array, is done so recursively. Commonly, the splitting will involve a function calling itself (recursion), as the most intuitive way to break down n elements until they are each in their own single-element lists.

This method makes Merge Sort highly consistant; predictable and reliable for sorting large arrays.

Due to the recursive nature of the Merge Sort, the complexity is a bit more mathematically involved (meaning it's complexity is unlikely to be involved in any programming interviews, unless they're maths focussed). Without getting into its depth, we can just say ***the time complexity for Merge Sort is O(nlogn)***.

Because Merge Sort breaks an array down into individual components, and sorts and builds them up from there, ***it is not adaptive***, as a nearly sorted list will be broken down onto just as many pieces as an unsorted one, and solved in likely nearly as many sorts and checks.

Because Merge Sort immediately breaks the input array down into sub-arrays, and builds it back up again, the original array is never modified. This means ***Merge Sort does *not* sort in-place***. For the same reason, breaking an array down into individual sub-arrays for each element before building it back up, we can safely say ***the extra space Merge Sort uses is O(n)***.

***Merge Sort is stable***, because swaps are made only if elements are of different values; same values will retain their original positions in relation to one another.

### Version 1 - With Recursion (Common Implementation)

```python
def merge_sort(list_input):
    if len(list_input) <= 1:
        # Once the list is only 1 element, it is returned, instead of further recursion
        print(f"  3. {list_input} is only 1 element; returning")
        return list_input

    mid = len(list_input) // 2 # Find the middle point of the list, to divide the elements between
    print(f"1. Middle of list {list_input} is index {mid}")
    leftHalf = list_input[:mid]
    rightHalf = list_input[mid:]
    print(f" 2. leftHalf is {leftHalf} and rightHalf is {rightHalf}")
    
    sortedLeft = merge_sort(leftHalf)
    sortedRight = merge_sort(rightHalf)
    print(f"   4. SortedLeft is {sortedLeft}, sortedRight is {sortedRight}")

    return merge(sortedLeft, sortedRight)

def merge(left, right):
    print("    5. merge function called")
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])

    print(f"     6. merge function returning {result}")

    return result
```

```python
input_list = [15, 32, 26, 11, 36, 19, 42, 44, 14]

array = merge_sort(input_list)
print(array)
```

It should be noted that this version makes use of recursion; the merge_sort function calls itself when defining the sortedLeft and sortedRight sub-arrays, until recursion stops when it's detected the lists have a single element, returning the single-element list instead of further calling itself.

Now whilst this example is much more easily understood, as well as using less code lines, recursion can easily lead to big problems if you're not careful; without proper handling, recursive functions can cause an infinite feedback loop.

Now, this is not to say that this version is better or worse than the next version; most common implementations of Merge Sort rely on recursion, just that recursion is both very useful and very dangerous. If you want a version that uses no recursion, then you can take a look at the next version.

In many programming languages, the built-in sort function is actually a quick sort. Note, this doesn't include Python, which uses [Timsort](https://en.wikipedia.org/wiki/Timsort), derived from merge sort and insertion sort, designed to perform well on many kinds of real-world data.

Quick Sort is a divide-and-conquer algorithm; rather than being based on the length of an array, it's based on a pivot; one of the elements, selected from the list, where then the list is partioned, with every element smaller than the pivot moved to the left of the pivot, and every element larger than the pivot to the right of it.

These partitions, reach representing a sublist, are in turn partioned with a pivot, continuing to be partioned until the entire list is sorted.

***The basic steps of the Quick Sort algorithm are:***
1. Choose a value in the array to be the pivot element.
2. Order the rest of the array so that lower values than the pivot element are on the left, and higher values are on the right.
3. Swap the pivot element with the first element of the higher values so that the pivot element lands in between the lower and higher values.
4. Do the same operations (recursively) for the sub-arrays on the left and right side of the pivot element.

It is a common tactic is chosen pivot is either left-most, right-most, or centre element of the list/sub-list.

[15, 32, 26, 11, 36, 19, 42, 44, 14]

Say we pick the rightmost element, so the sublist is the entire list, and 14 happens to be the smallest; that pivot element will make its way to its correct position in the list during this iteration.
In the process, we must also ensure that all elelemtns left of the pivot are smaller, and everything right is later.

So we'd iterate from the elements from left to right, until we find something smaller than our 14 pivot; when we do, we swap that smaller element all the way to the left of the list, now the first element, and swapped out the first element into the position the smallest once occupied. So 15 gets swapped with 11, where 11 now rests as the smallest value.

[11, 32, 26, 15, 36, 19, 42, 44, 14]

Lets say there are no more elements smaller than 14, so this iteration ends with 14 being swapped into the position after the element we found that was smaller than 14; so swapped with the element after 11, which is 32. So 14 is now in the second position; thus, as foretold, our pivot element is now in its rightful place.

[15, 14, 26, 11, 36, 19, 42, 44, 32]

When implementing Quick Sort, there are two functions that need to be defined:
- A "partition" function, which is given a subarray as its input. It picks one of the elements as a pivot, sorting elements smaller than the pivot to the left and larger to the right; it ensures the pivot element moves to its rightful place in that subarray.
- A "quicksort" function, which makes all the recursive calls to the sublists. It is the function which invokes the calls to the partition function and ensures that each of the partitions are sorted.

Due to the recursive nature of the Quick Sort, the complexity is a bit more mathematically involved (meaning it's complexity is unlikely to be involved in any programming interviews, unless they're maths focussed). Without getting into its depth, we can just say ***the average time complexity for Qucick Sort is O(nlogn)*** (worst case of O(n^2).

Depending on the implementations and optimisations used, ***Quick Sort will on average make O(nlogn) comparisons***.

```python
def quick_sort(input_list, low=0, high=None):
    print(f"(quick_sort) list = {input_list}, low = {low}, high = {high}")
    if high is None:
        # Make pivot the rightmost value
        high = len(input_list) - 1
        print(f"Set initial pivot to element at index {high}; {input_list[high]}")

    if low < high:
        pivot_index = partition(input_list, low, high)
        quick_sort(input_list, low, pivot_index-1)
        quick_sort(input_list, pivot_index+1, high)

def partition(input_list, low, high):
    print(f"(partition) list = {input_list}, low = {low}, high = {high}")
    print(f"partion is {input_list[low:high+1]}")
    pivot = input_list[high]
    print(f"Pivot is {pivot}")
    i = low - 1

    for j in range(low, high):
        print(f"Is {input_list[j]} less than or equal to {pivot}? {input_list[j] <= pivot}")
        if input_list[j] <= pivot:
            # If we find an element that's less than the pivot, we swap it into the leftmost place we haven't already sorted
            # So, for every element we find in an iteration that's less than the pivot, i increases; because when we finish the iteration
            # the pivot will swap with what element is in index i+1
            # For example, in first iteration, we find 11 < 14; swap 11 into the start, increase 1; nothing else < 14, so we swap 14
            # into the next avaialble slot; at i+1
            i += 1
            print(f"Swap elements {input_list[i]} and {input_list[j]} around")
            print(f"Set {input_list[j]} as smallest value at index {i}")
            input_list[i], input_list[j] = input_list[j], input_list[i]

    print(f"Partition ended; swap pivot {pivot} with {input_list[i+1]} at index {i+1}")
    print(f"Set {pivot} as smallest value at index {i+1}")
    input_list[i+1], input_list[high] = input_list[high], input_list[i+1]
    return i+1
```
