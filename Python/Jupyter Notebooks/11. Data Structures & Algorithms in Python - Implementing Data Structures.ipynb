### O(n) Operations

```python
# O(n) refers to operations where the number of iterations varies linearly with the input
```

```python
def check_prime1(number):
    num_iterations = 0
    for i in range(2, number):
        num_iterations += 1

        if number % i == 0:
            print(f"{number} is not a prime number \nTotal number of iterations = {num_iterations}")
            return

    print(f"{number} is a prime number \nTotal number of iterations = {num_iterations}")
```

```python
# By changing the value of the argument, the number of iterations changes too
```

```python
# We've seen that, worst case, this function needs as many iterations as the number in order to execute; if our input is n, number of iterations is n
# This also means its complexity can be represented as O(n)
```

```python
# There is a more efficient way to check for a prime number however; we can cut the potential complexity in half
# We know that any number can't be divisible by more than half of itself, save for itself, making it a prime number
def check_prime2(number):
    num_iterations = 0
    mid_point = int(number / 2)
    for i in range(2, mid_point):
        num_iterations += 1

        if number % i == 0:
            print(f"{number} is not a prime number \nTotal number of iterations = {num_iterations}")
            return

    print(f"{number} is a prime number \nTotal number of iterations = {num_iterations}")
```

```python
check_prime2(1845)
```

```python
check_prime2(1847)
```

**It's worth noting that even though we've managed to cut this function's complexity in half, theoretically n/2, Big O notation still considers this O(n). This is because reduction, dividing by two; a constant, only affects the function's growth rate by a constant amount. As Big O notation doesn't care about constants, it's ignored.**

```python
# We've covered functions which take a single integer, but what about functions with lists as the input argument
```

```python
def find_maxval(num_list):
    maxval = num_list[0]
    num_iterations = 0

    for i in range(len(num_list)):
        num_iterations += 1

        if maxval < num_list[i]:
            maxval = num_list[i]

    print(f"Maximum value of the list = {maxval} \nTotal number of iterations = {num_iterations}")
```

```python
find_maxval([510, 635, 49, 102])
# We see the number of iterations is directly proportional with the length of the list; n iterations for n-long list
```

```python
find_maxval([510, 635, 49, 102, 500, 629, 400, 325, 80, 723])
# Thus, this function's complexity can be represented by O(n)
```

```python
# NOTE!! This function has a type hint, which shows to users who try to write out this function that we're expecting an int to be provided
def find_factorial(number:int):
    fact = 1
    num_iterations = 0

    # This will check for every iteration
    if(number < 0 or type(number) != int):
        print("Invalid number")
        return

    for i in range(1, number+1):
        fact = fact * i
        num_iterations += 1

    print(f"The factorial of {number} = {fact} \nTotal number of iterations = {num_iterations}")
```

```python
# What about multiple loops in a single function?
def two_for_loops(number):
    total = 0
    num_iterations = 0

    for i in range(number):
        num_iterations += 1
        total = total + i

    for i in range(100):
        num_iterations += 1
        total = total + i

    print(f"Total iterations are {num_iterations}")
```

```python
# We see the buttom loop, unchanged by the input, will always run 100 iterations; this means the function's complexity is n + 100
# However, this is a good time to introduce one of the theories behind Big O notation; lower order terms and constants *don't* matter.
```

When we consider the complexity of any task, we assume n is a very, *very* large number, and that any lower order terms or constants do not matter while expressing complexity. This means that the constant 100, which would be irrelevant if n was 10^15, is ignored. Thus, we are still left with a complexity of O(n).

Even if the 2nd loop ran n times, leading the function to theoretically have a complexity of 2n, Big O notation doesn't 'care'. "Big-O notation doesn't care about constants because big-O notation only describes the long-term growth rate of functions, rather than their absolute magnitudes. Multiplying a function by a constant only influences its growth rate by a constant amount, so linear functions still grow linearly, logarithmic functions still grow logarithmically, exponential functions still grow exponentially, etc. Since these categories aren't affected by constants, it doesn't matter that we drop the constants."

A fantastic link describing why Big O notation discards lower order terms and constants can be found [here](https://stackoverflow.com/a/22188943)

### O(n^2) Operations

```python
# In the previous topic, we had two for loops in a single function, independant of one another, that still only raised the complexity to O(n)
# But what if the for loops are nested? For every n iterations of the first loop, we run n iterations of the second.
def print_pairs(number_list):
    num_iterations = 0
    n = len(number_list)

    for i in range(n):
        for j in range(n):
            print(number_list[i], number_list[j])
            num_iterations += 1

    print(f"Total iterations are {num_iterations}")
```

```python
print_pairs(["A", "B", "C"])
```

```python
print_pairs(["A", "B", "C", "D", "E", "F"])
```

```python
# 3 values in the list led to 9 iterations; 6 in the list led to 36; n*n iterations, or n^2
# We've created one of the most obvious examples of a function of complexity O(n^2)
```

```python
def find_prime(lower, upper):
    num_iterations = 0

    for num in range(lower, upper):
        for i in range(2, int(num / 2)):
            num_iterations += 1
            if(num % i) == 0:
                break
            else:
                print(num)
    print(f"Total iterations are {num_iterations}")
```

```python
find_prime(0,10)
```

```python
find_prime(10, 20)
```

```python
find_prime(10, 2000)
```

```python
# Here we can see, with an example on the larger side, that the complexity is O(n^2); the complexity grows exponentially as n does
```

## Defining a Custom Queue

In order to work with a queue, in this topic we will write a class for our own queue, in which we will store the elements of a queue in a Python list.

```python
# This is how we define a class; a blueprint to create a MyQueue object; each of these is a function of the object
class MyQueue:
    def __init__(self):
        """Create a new queue."""
        self.items=[]
    def is_empty(self):
        """Return true if queue is empty."""
        return len(self.items) == 0
    def enqueue(self, item):
        """Add a new element to the end of the queue."""
        self.items.append(item)
    def dequeue(self):
        """Remove an element from the beginning of the queue."""
        return self.items.pop(0)
    def size(self):
        """Return the size of the queue."""
        return len(self.items)
    def peek(self):
        """Return the first element of the queue."""
        if self.is_empty():
            raise Exception("Nothing to peek")
        return self.items[0]
```

## Defining a Custom Stack

Whilst we saw we can comfortable use a Python list as a stack, it may be desirable to extend the behaviour of a regular Python list in order to implement additional operations with a stack.

```python
class Stack:
    def __init__(self):
        """Create a new stack."""
        self.stack = []
    def push(self, data):
        """Add an element to the stack."""
        self.stack.append(data)
    def pop(self):
        """Remove the top element of the stack."""
        if self.is_empty():
            raise Exception("Nothing to pop")
        return self.stack.pop(len(self.stack) - 1) # This pops the last element and returns it
    def peek(self):
        """Return the top element of the stack"""
        if self.is_empty():
            raise Exception("Nothing to peek")
        return self.stack[len(self.stack) - 1]
    def is_empty(self):
        """Return true if stack is empty."""
        return len(self.stack) == 0
    def size(self):
        """Return size of the stack"""
        return len(self.stack)
```

### Defining Linked List Node

```python
# dataval contains the data of the node, whilst nextval contains the pointer to the next node in the linked list
class Node:
    """A node in a singly-linked list."""
    def __init__(self, dataval=None, nextval=None):
        self.dataval = dataval
        self.nextval = nextval
    # __repr__ stands for the "representation" of the object
    def __repr__(self):
        """Print out the conents of the node."""
        return repr(self.dataval)
```

### TODO!! Defining Linked List (Insert, Search, Delete, Reverse, etc. Operations)

```python
# TODO!! Leave comments to make sense of this beast, especially the prepend and append functions
class LinkedList:
    def __init__(self):
        """Creating a new singly-linked list."""
        # Any implementation of a linked list needs a variable which points to the head node of the linked list
        self.head = None
    
    def __repr__(self):
        """Creating a string representation of the data in a list."""
        nodes = []
        curr = self.head

        # curr will jump across the linked list, appending the repr of each node as it does, until curr = null
        while curr:
            nodes.append(repr(curr))
            curr = curr.nextval
        return '[' + '->'.join(nodes) + ']'
        
    def prepend(self, dataval):
        """Insert a new element at the beginning of the list."""
        self.head = Node(dataval=dataval, nextval=self.head)
        
    def append(self, dataval):
        """Insert a new element at the end of the list."""
        if not self.head:
            self.head = Node(dataval=dataval) # Don't need to set nextval if not the head, as it defaults to None
            return
        curr = self.head
        while curr.nextval:
            curr = curr.nextval
        curr.nextval = Node(dataval=dataval)
    
    def add_after(self, middle_dataval, dataval):
        """Insert a new element after the node with middle_dataval."""
        if middle_dataval is None:
            print("Data to insert after not specified")
            return
            
        curr = self.head
        
        while curr and curr.dataval != middle_dataval:
            curr = curr.nextval
        new_node = Node(dataval=dataval)

        new_node.nextval = curr.nextval
        curr.nextval = new_node
    def find(self, data):
        """Search for the first element with `dataval` matching `data`. 
        Returns the element or `None` if not found."""
        curr = self.head

        while curr and curr.dataval != data:
            curr = curr.nextval

        return curr

    def remove(self, data):
        """Remove the first occurrence of `data` in the list."""
        curr = self.head
        prev = None

        while curr and curr.dataval != data:
            prev = curr
            curr - curr.nextval

        if prev is None:
            self.head = curr.nextval
        elif curr:
            prev.nextval = curr.nextval
            curr.nextval = None

    def reverse(self):
        """Reverse the list in-place."""
        curr = self.head

        prev_node = None
        next_node = None

        while curr:
            nextval = curr.nextval
            curr.nextval = prev_node

            prev_node = curr
            
            curr = nextval
            
        self.head = prev_node

    def reverse_recursive(self):
        """Reverse the list in place using recursion."""
        def recursion(curr, prev) :
            if not curr:
                return prev

            nextval = curr.nextval
            curr.nextval = prev

            prev = curr
            curr = nextval

            return recursion(curr, prev)

        self.head = recursion(curr=self.head, prev=None)

    def count_nodes(self):
        """Count the number of nodes in the linked list."""
        if (self.head == None):
            return 0
        else:
            curr = self.head
            count = 0
            while (curr != None) :
                curr = curr.nextval
                count += 1
                
        return count
```
