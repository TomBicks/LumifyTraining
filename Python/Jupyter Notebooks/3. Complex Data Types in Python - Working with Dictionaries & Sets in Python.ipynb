# Complex Data Types in Python: Working with Dictionaries & Sets in Python

## Introducing Dictionaries

```python
# A dictionary is created with '{}'
empty_dict = {}
empty_dict
```

```python
# A dictionary is a collection of key:value pairs
favourite_colours = {"Tom":"Green",
                     "Steph":"Blue",
                     "Jaydan":"Red"}
favourite_colours
```

```python
# We can use the key to look up value it's tied to in a dictionary
favourite_colours["Tom"]
```

```python
# However the keys need to be *exactly* correct to work, or they will throw an error if the key doesn't exist in the dictionary
favourite_colours["tom"]
```

```python
# Dictionary keys should be unique, else they can't be used as a unique identifier
# If the same key is used twice, rather than an error, Python will only take the last value of the key (the latest time that key is used in the dict.)
duplicate_keys = {1:4, 2:17, 3:50, 3:75}
print(duplicate_keys.keys())
print(duplicate_keys)
# Notice 3 only shows up once, with a value of 75; the last instance of 3 being used as a key
```

```python
# We can use the 'in' keyword to check whether a particular key is present in the Dictionary; this way we can check without throwing errors
"Samuel" in favourite_colours
```

```python
"Tom" in favourite_colours
```

```python
# Keys, whilst most commonly strings or integers, can be other data types
mixed_dict = {True:"Daniel",
              7.3:[1, 2, 3],
              "Jacob":False}
mixed_dict
```

```python
print(mixed_dict[True])
print(mixed_dict[7.3])
print(mixed_dict["Jacob"])
```

```python
# An alternative to retrieving data, instead of using dict[], is using the .get function
print(favourite_colours["Steph"])
print(favourite_colours.get("Steph"))
```

```python
# To delete a key:value pair, usually after its prupose is done or outdated, we use the 'del' keyword
del favourite_colours["Tom"]
favourite_colours
```

## Nesting Complex Data Types within Dictionaries

```python
# We can assign values to a dictionary by providing a key to assign it to to assign new key:value pairs
fruit_qty = {}
fruit_qty
```

```python
fruit_qty["Banana"] = 50
fruit_qty["Apple"] = 71
fruit_qty["Plum"] =18
```

```python
fruit_qty
```

```python
# Values inside a dictionary can be complex data types, such as lists here
fruit_qty_consumed = {'Banana':[12, 20, 8],
                     'Apple':[40, 25, 37],
                     'Plum':[2, 0, 18]
                    }
fruit_qty_consumed
```

```python
# And of course we can access the values at indexes of these list values
fruit_qty_consumed['Banana'][1]
```

```python
# And we can assign them the same way
fruit_qty_consumed['Banana'][2] = 80
fruit_qty_consumed['Banana']
```

```python
# Values within dictionaries can themselves also be dictionaries
fruit_qty_consumed_days = {'Banana':{'Fri':12, 'Sat':20, 'Sun':8},
                     'Apple':{'Fri':40, 'Sat':25, 'Sun':37},
                     'Plum':{'Fri':2, 'Sat':0, 'Sun':18}
                    }
fruit_qty_consumed_days
```

```python
fruit_qty_consumed_days['Plum']['Fri']
```

## Invoking Functions on Dictionaries

```python
bike_details = {"bike_owner" : "James Smith" ,
                "bike_model" : "Ducati Monster 1200" ,
                "bike_price" : 28140 ,
                "engine_displacement" : 1197}
bike_details
```

```python
# Length refers to how many key:value pairs there are in the context of a dictionary
len(bike_details)
```

```python
# Sorts the keys in a lexographical order
sorted(bike_details)
```

```python
# NOTE!! This is how you reverse a sorted function too
sorted(bike_details, reverse=True)
```

```python
# NOTE!! Whilst you *can* sort the values of a dictionary, if there are differeing data types like strings and ints, it will throw an error
# This is because Python can't determinem for example, whether "red" > 3
sorted(bike_details.values())
```

```python
# .items returns a list of key:value pairs in the form of tuples
bike_details.items()
```

```python
# You can use the .copy function to create a copy of a dictionary; these are deep copies, with changes made to the copy not affecting the original
copy_bike_details = bike_details.copy()
copy_bike_details
```

```python
del copy_bike_details['bike_owner']
bike_details
```

```python
copy_bike_details
```

```python
# We can also use .pop to remove a key:value pair from a dictionary
# NOTE!! .pop returns the value that was removed from the complex data
popped = copy_bike_details.pop('bike_price')
print(popped)
print(copy_bike_details)
```

### Update Function

```python
# Dictionaries support update functions
dict_age = {'Ethan':54, 'Sofia':50}
new_dict_age = {'Ethan':55, 'Harper':58}
```

## Introducing Sets

```python
# A commonly used complex data type, sets are collections of elements where the elements are unique
# They contain no duplicates, nor do they have an intrinsic order (meaning the order cannot be relied upon)
# Sets are initialised using using '{}', just like dictionaries, but have no key:value pairs, to differentiate them
set_string = {'Emma', 'Olivia', 'Ava', 'Mia'}
set_string
```

```python
print(set_string)
# Notice the order is different? Again, there is *no* intrinsic order
```

```python
# To quickly create an empty set, you use 'set()', to differentiate it from dictionaries
empty_set = set()
empty_set
```

```python
# Sets can have different data types, again provided the elements are unique
# NOTE!! Tuples are the only complex data type gone over so far which are immutable
mixed_set = {'Emma', 7.3, False, (1, 2, 3)}
mixed_set
```

```python
# Lists however *cannot* be part of a set; this is because the elements in a set can only be immutable values; lists can be changed and thus mutable
set_list = {1, 2, 3, [4, 5, 6]}
```

```python
# Likewise, dictionaries are mutable, and cannot be part of a set
set_dict = {1, 2, 3, {'Ted':7, 'Bob':8}}
```

```python
# Sets elements must be unique, meaning duplicate entries are elminated
student_set = {'Emma', 'Jane', 'Maria', 'Emma', "Jane"}
student_set
```

```python
# To add an element
student_set.add('Anna')
student_set
```

```python
# Max can be used on a set of strings, but will just return the string with the most characters
max(student_set)
```

```python
# We can use .remove to remove an element
student_set.remove('Jane')
student_set
```

```python
# We can also use .discard to remove an element
student_set.remove('Maria')
student_set
```

```python
# The difference between .remove and .discard is, .remove will throw an error if it can't find what to remove, whereas .discard will do nothing
student_set.discard('Maria')
```

```python
student_set.remove('Maria')
```

```python
# To clear, .clear works here too
student_set.clear()
student_set
```

## Performing Set Operations

```python
# To combine sets together, we use the .union function
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
set3 = {8, 9, 10, 11, 12}
```

```python
set1.union(set2)
# Notice the duplicate entries, 8 and 9, are now singular?
```

```python
# Union accepts multipl arguments to combine multiple sets
number_union = set1.union(set2, set3)
number_union
```

```python
# To return the elements that are present in multiple sets, we use the .intersection function
set2.intersection(set1)
```

```python
# To obtain the elements unique to a set compared to other sets, we use the .difference function
# NOTE!! This doesn't include the unique elements from the argument set, only the one calling the function
set2.difference(set1)
```

```python
# If we want to update a set to, for example, be updated with the results of an intersection, we call .intersection_update
set1.intersection_update(set2)
print(set1)
print(set2)
# Notice set 1 is now updated, where before it was unchanged
```

```python
# Likewise with difference_update
set2.difference_update(set3)
set2
```

```python
# The function .isdisjoint checks whether two sets are 'disjoints'; whether they contain common elements or not
print({1,2,3}.isdisjoint({4,5,6})) # Disjointed - shares no elements
print({1,2,3}.isdisjoint({3,4,5})) # Not Disjointed - shares a 3
```

```python
print(set1)
print(set2)
print(set3)
```

```python
# Set1 and Set2 share common elements, so they are not disjointed
set1.isdisjoint(set2)
```

```python
# Set1 and Set3 share no common elements, so they are disjointed
set1.isdisjoint(set3)
```

## Working with Nested Lists

```python
# Lists can contain any data type, including other lists. Nested lists are also known as matrices
car_matrix = [['Hennessey Venom GT', 1244 ],
              ['SSC Ultimate Aero', 1287],
              ['Zenvo ST1', 1100]]
car_matrix
```

```python
print(car_matrix[1])
print(car_matrix[1][0])
```

```python
# the len function (length) will only return the length of the outermost list
len(car_matrix)
```

```python
# However, we can still get the length of the list inside by retrieving the nested list within car_matrix
len(car_matrix[0])
```

```python
# As with all lists, if the index is out of range, even for the nested list, an error will be thrown
car_matrix[0][2] # Index 0 is correct, but 2 is out of range
```

```python
# Slicing operations are possible iwht nested lists as well
car_matrix[2][0:2]
```

```python
car_matrix[1][:-1]
```

## Performing List Conversions

```python
# We can perform conversions of lists into other complex data types, such as a tuple
my_list = ['Leo', 18, 'Aaron', 25, 'Easton', 34, 'Jordan', 30]
```

```python
# We can convert the list into a tuple
tuple(my_list)
# Notice that, due to the lack of nesting within the original list, this tuple is one-dimensional
```

```python
# Here is a list of nested lists
my_list2 = [['Leo', 18], 
            ['Aaron', 25], 
            ['Easton', 34], 
            ['Jordan', 30]]
my_list2
```

```python
# Converting this nested list to a tuple will convert the outer layer into a tuple, whilst the inner fields are still lists
tuple(my_list2)
```

```python
# Here is another list of nested lists
my_list3 = [[1, 'Stephen', 2, 'Reg'], 
            [3, 'Richard', 4, 'Anne']]
my_list3
# We can think of this list of lists as a list in two-dimensions (hence matrices)
```

```python
# Converting this list to a dictionary does not work as it is now, not sure how to sort keys from values
dict(my_list3)
# Notice the error? Length of 4 refers to the nested lists.
# What this means is that, if the nested lists are only of 2 elements, Python assumes the first element is the key and the second is the value
```

```python
my_list3 = [[1, 'Stephen'], 
            [2, 'Reg'], 
            [3, 'Richard'], 
            [4, 'Anne']]
my_list3
```

```python
# Now the conversion should work
dict(my_list3)
```

```python
# However, what if we only have one nested list now viable?
my_list3 = [[1, 'Stephen'], 
            [2, 'Reg'], 
            [3, 'Richard'], 
            [4, 'Anne', 'Angela']]
my_list3
```

```python
dict(my_list3)
```

```python
# We need the last list to be a list with two elements. We already know elements in a list can be in of themselves a list. 
# So we put the values that don't fit into a list as the second value
my_list3 = [[1, 'Stephen'], 
            [2, 'Reg'], 
            [3, 'Richard'], 
            [4, ['Anne', 'Angela']]]
my_list3
```

```python
dict(my_list3)
# Now the last value of the dictionary is a list
```

```python
# If we take the matrix from the last topic, we see it can easily be converted into a dictionary
car_matrix = [['Hennessey Venom GT', 1244 ],
              ['SSC Ultimate Aero', 1287],
              ['Zenvo ST1', 1100]]
car_matrix
```

```python
car_matrix_dict = dict(car_matrix)
car_matrix_dict
```

```python
# If we then edit the car_matrix, we still see however, that the dictionary created from it remains unchanged; it no longer refers to the original list
car_matrix.clear()
car_matrix
```

```python
car_matrix_dict
```

```python
# If we try to go backwards and convert the dictionary into a list, the results are unexpected
car_names = list(car_matrix_dict)
car_names
# We see only the key names have been carried over; not the values
```

```python
# To get the values, we need to specify it when converting
car_power = list(car_matrix_dict.values())
car_power
```
